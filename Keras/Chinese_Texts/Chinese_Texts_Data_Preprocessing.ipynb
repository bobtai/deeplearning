```python
# å°‡ df è½‰æ›æˆ nd_array
nd_array = data_frame.values
```

```python
# å–å‡º features å’Œ labelsï¼Œä¸¦è½‰æˆæ¨™æº– list
features = nd_array[:,0].tolist() # å–ç¬¬ 0 æ¬„ä½œç‚º features
labels = nd_array[:,1].tolist() # å–ç¬¬ 1 æ¬„ä½œç‚º labels
```

```python
# raw data
features[2053]
```

```python
# å°‡è¨Šæ¯ä¸­çš„ä¸­æ–‡å­—éæ¿¾å‡ºä¾†
import re
def filter_chinese_letter(contexts):
    # \u4E00-\u9FA5 Chinese unicode range
    # a-zA-Z English range
    filter_re = re.compile('[^\u4E00-\u9FA5]') # non Chinese
    i = 0
    for context in contexts:
        contexts[i] = filter_re.sub('', str(context)) # remove all non Chinese letters                   
        i = i + 1
    return contexts
features = filter_chinese_letter(features)
```

```python
# processed data
features[2053]
```

```python
from keras.preprocessing.text import Tokenizer
token = Tokenizer(num_words=1000, # å°‡æœ€å¸¸å‡ºç¾çš„ 1000 å­—ä½œç‚º Token
                  char_level=True) # ä»¥å–®å€‹å­—åˆ† Token
token.fit_on_texts(features)
token.word_index
```

```python
# æ ¹æ“šå­—å…¸ï¼Œå°‡è¨Šæ¯æ–‡å­—è½‰æˆæ•¸å­— listï¼Œä¸åœ¨å­—å…¸ä¸­çš„æ–‡å­—ä¸è½‰æ›ã€‚
train_seq = token.texts_to_sequences(features)
train_seq[0]
```

```python
# æ¯å€‹è¨Šæ¯æ–‡å­—çš„æ•¸å­— list é•·åº¦ä¸åŒï¼Œè¦é€²è¡Œæ·±åº¦å­¸ç¿’ï¼Œé•·åº¦å¿…é ˆå›ºå®šã€‚
from keras.preprocessing import sequence
train_features = sequence.pad_sequences(train_seq, maxlen=20)
train_features[0]
```

```python
# å°‡ label è½‰æ›æˆ onehot æ ¼å¼
from keras.utils import np_utils
train_labels = np_utils.to_categorical(labels)
train_labels[0]
```
